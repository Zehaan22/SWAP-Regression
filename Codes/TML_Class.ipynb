```python
import numpy as np

# Generating Random Data
n = 10
X = 10* np.random.random(n) - 5 # (0,1) --> (-5,5)
Y = 4*X + 5 + np.random.normal(scale= 5,size = n)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(
    X,
    Y
)
```

```python
X_train = []
y_train = []
for i in range(len(X)):
    X_train.append((1,X[i]))
    y_train.append(Y[i])
    
X = np.matrix(X_train)
y = np.matrix(y_train)
```

```python
beta = (X.T*X).I*X.T*y.T
beta
```

```python
y_pred = X*beta

plt.figure()
plt.scatter(
    np.array(X[:,1]),
    np.array(y),
    c = "blue"
)
plt.scatter(
    np.array(X[:,1]),
    np.array(y_pred),
    c = "red"
)
```
