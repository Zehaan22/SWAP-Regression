**Importing all the necessary modules including SimpleRegression**

Our file SimpleRegression has only onle function, called generate_parameters that gives b0 and b1 as the coefiicients for the regression equation.

```python
import matplotlib.pyplot as plt
import csv
import os
import numpy as np
from simpleRegression import generate_parameters
from multipleRegression import getBeta
```

```python
cwd = os.getcwd()

filename = "Data Sets/death.csv"
path = os.path.join(cwd, filename)
```

```python
X = []
Y = []

with open(path, "r+") as file:
    reader = csv.reader(file)
    count = 0
    for i in reader:
        if (count >= 1 and count <= 200):
            X.append([float(i[3]),float(i[4]),float(i[5])])
            Y.append(float(i[1]))
        count += 1
```

```python
plt.plot(X, Y, 'o')

X = np.matrix(X)
y = np.matrix(Y).T
B = getBeta(y,X)
```

### Multiple Regression Model Testing

```python
X_ = np.c_[np.ones(X.shape[0]), X]
yCap = X_*B
plt.plot(yCap,Y, "o")
print(B)
```

**Plotting the graph**

The *orange* line is the line where Y is regressed on X. We observe the expected match accuracyh and prediction score.

The *green* line is the line where X is regressed on Y. We observe that it does not depict the data accurately.

```python
plt.plot(X, Y, 'o')
b0, b1 = generate_parameters(X, Y)
Y_prime = []
for i in X:
    Y_prime.append(b0 + b1*i)
plt.plot(X, Y_prime)
print(b0,b1)

b0, b1 = generate_parameters(Y, X)
Y_prime = []
for i in X:
    Y_prime.append(b0 + b1*i)
plt.plot(X, Y_prime)
print(b0,b1)

plt.show()
```

NOW, we move on to an interesting Data Set:

THE NY Stock Exchange.

```python
cwd = os.getcwd()

filename = "Data Sets/fundamentals.csv"
path = os.path.join(cwd, filename)

Parameters = dict()

with open(path, "r+") as file:
    reader = csv.reader(file)
    count = 0
    attrs = reader.__next__()
    attrs[0] = "Index"
    
    for i in attrs:
        Parameters[i] = []
    
    for i in reader:
        for j in range(len(attrs)): 
            try:
                if (count >= 1 and count <= 500):
                    Parameters[attrs[j]].append(float(i[j]))
                else:
                    break
                    
            except:
                Parameters[attrs[j]].append((i[j]))
        count += 1
```

```python
X = Parameters["Accounts Payable"]
Y = Parameters["Earnings Before Interest and Tax"]

plt.plot(X, Y, 'o')
b0, b1 = generate_parameters(X, Y)
Y_prime = []
for i in X:
    Y_prime.append(b0 + b1*i)
plt.plot(X, Y_prime)
print(b0,b1)

b0, b1 = generate_parameters(Y, X)
Y_prime = []
for i in X:
    Y_prime.append(b0 + b1*i)
plt.plot(X, Y_prime)
print(b0,b1)

plt.show()
```
