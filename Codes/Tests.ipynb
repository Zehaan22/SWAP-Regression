**Importing all the necessary modules including SimpleRegression**

Our file SimpleRegression has only onle function, called generate_parameters that gives b0 and b1 as the coefiicients for the regression equation.

```python
import matplotlib.pyplot as plt
import csv
import os
from SimpleRegression import generate_parameters
```

```python
cwd = os.getcwd()

filename = "Data Sets/death.csv"
path = os.path.join(cwd, filename)
```

```python
X = []
Y = []

with open(path, "r+") as file:
    reader = csv.reader(file)
    count = 0
    for i in reader:
        if (count >= 1 and count <= 200):
            X.append(float(i[3]))
            Y.append(float(i[4]))
        count += 1
```

**Plotting the graph**

The *orange* line is the line where Y is regressed on X. We observe the expected match accuracyh and prediction score.

The *green* line is the line where X is regressed on Y. We observe that it does not depict the data accurately.

```python
plt.plot(X, Y, 'o')
b0, b1 = generate_parameters(X, Y)
Y_prime = []
for i in X:
    Y_prime.append(b0 + b1*i)
plt.plot(X, Y_prime)
print(b0,b1)

b0, b1 = generate_parameters(Y, X)
Y_prime = []
for i in X:
    Y_prime.append(b0 + b1*i)
plt.plot(X, Y_prime)
print(b0,b1)

plt.show()
```
